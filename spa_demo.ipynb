## Detailed Access and Usage Guide for the Demonstration

### Medical Costs Prediction SPA
To access and use the Medical Costs Prediction Single Page Application:

1. **Start the Backend API:**
   - Navigate to the 'Medical Costs Project' directory.
   - Find the `api.py` file, which contains the Flask API code.
   - Open a terminal or command prompt window.
   - Navigate to the directory containing `api.py` using the `cd` command.
   - Run the API by typing `python api.py` into the terminal and pressing Enter. This will start the Flask server, which needs to be running for the SPA to interact with the machine learning model.

2. **Launching the Application:**
   - With the Flask API running, open a new terminal or command prompt window without closing the first one.
   - In the new terminal window, navigate to the directory containing the `SPA.html` file if it's different from where `api.py` is located.
   - Type `python -m http.server` and press Enter. This command starts a simple HTTP server to serve the SPA file.
   - Note the port number given by the HTTP server (usually `8000`).

3. **Accessing the Application:**
   - Open a web browser.
   - Type `localhost:8000/SPA.html` in the browser's address bar, replacing `8000` with the actual port number if different.
   - Press Enter to open the SPA.

4. **Using the Application:**
   - The SPA will present a form where you can input features such as age, BMI, smoking status, and so on.
   - Fill in the fields with the data for which you want to predict insurance costs.
   - Click the 'Predict' button to receive the cost prediction based on the Random Forest model.

Ensure that both the Flask API and the HTTP server are running concurrently; the API serves the model predictions, while the HTTP server serves the SPA to your browser.

By following these steps, even users unfamiliar with web server setup or machine learning model deployment can successfully launch and interact with the applications. These detailed instructions aim to make the powerful capabilities of machine learning accessible and user-friendly.
